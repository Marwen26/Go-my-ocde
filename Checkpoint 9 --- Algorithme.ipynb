```python
# Question 1  : Binary Serch

def B_Serh  (My_list , l , h , v):
    if h >= l :
        mid = int(l+(h-l)/2)
        if My_list[mid] == v:
            return "True"
        elif My_list[mid] > v :
            return B_Serh(My_list , l , mid-1 , v)
        else :
            return B_Serh(My_list , mid+1 , h , v)
    else:
        return "False"

My_list = [ 1 , 2 , 5 , 6 , 8 , 10 , 12 , 14 , 55]
l = 0                  # low index of the list
h = len(My_list)-1     # high index of the lsit
v = 2                 # value serch
B_Serh(My_list , l , h , v)
```

```python
# Question 2 : 'a' to the power 'b'
a = int(input("a = "))
b = int(input("b = "))
def power (a,b) :
    if  a == 0 :
        return 0
    elif b == 0 :
        return 1
    elif b == 1 :
        return a
    else :
        return a*power(a,b-1)
print(power(a,b))        
```

```python
# Question 3 :  Bubble Sort

def B_sort(L):
    for p in range(len(L)-1, 0 , -1 ):               #range (list   start , stop , step)
        for i in range(p):
            if L[i] > L[i+1]:
                T = L[i]
                L[i] = L[i+1]
                L[i+1] = T
    return L
L = [ 29 , 13 , 22 , 37 , 52 , 49 , 46 , 71 , 56 ]
print(B_sort(L))
```

```python
# Question 4 :  Merg Sort

def M_sort(L):
    
    if len(L)>1:                                  
        l = len(L)//2                             
        left_L = L[:l]                            # left list
        right_L = L[l:]                           # right list

        M_sort(left_L)
        M_sort(right_L)
        
        i = 0                                     # i : iterators for the left list (left_L)
        j = 0                                     # j : iterators for the right list (right_L)
        
        k = 0                                     # k : for the main list (L)
        
        while i < len(left_L) and j < len(right_L):
            if left_L[i] < right_L[j]:
                L[k] = left_L[i]
                i +=1
            else :
                L[k] = right_L[j]
                j += 1
            k += 1
            
        while i < len(left_L):
            L[k] = left_L[i]
            i += 1
            k += 1
        while j < len(right_L):
            L[k] = right_L[j]
            j +=1
            k +=1
L = [29,13,22,37,52,49,46,71,56]
M_sort(L)
print(L) 
```

```python
# Question 5 :  Quick Sort:

# function partion : divide the main list to 2 list 

def partion(L,start,end):
    pv = L[start]                   # P : pivot
    low = start + 1
    high = end
    
    while True :
        while low <= high and L[high] >= pv:
            high = high - 1 
            
        while low <= high and L[low] <= pv:
            low = low + 1
        if low <= high:
            L[low],L[high] = L[high],L[low]
        else :
            break
    L[start],L[high] = L[high],L[start]
    return high

# Quick sort fuction :

def Q_Sort(L,start,end):
    if start >= end:
        return
    Par = partion(L,start,end)
    Q_Sort(L,start,Par-1)
    Q_Sort(L,Par+1,end)

L = [29,13,22,37,52,49,46,71,56]
Q_Sort(L,0,len(L)-1)
print(L)          
```
