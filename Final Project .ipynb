```python
# 1 : Take the two names from the user 
# 2 : Remove the common characters betwins the tow names and get the count of the characters that are left
# 3 : Take FLAMES list 
# 4 : Removing list elment using the count we got
# 5 : The letter which lasts the process is the result



# 1 : Take the two names from the user :

name_1 = input("Enter your name : ")
name_1 = name_1.lower()     # lower : convert the characters to lowercase if the user enter it uppercase 
# if condtion to chek if the name contain number or not :
if not name_1.isalpha() :
    print("Your name must contain letters only , Check and enter it again")
    name_1 = input("Enter your name : ")
name_2 = input("Enter your crash name : ")
name_2 = name_2.lower()
if not name_2.isalpha():
    print("Your crach name must contain letters only , Check and enter it again")
    name_2 = input("Enter your crach name : ")

L1 = []
L2 = []

def Falme(name_1 , name_2):
    
    for i in range (len(name_1)): 
        L1.append(name_1[i])
    for j in range (len(name_2)):
        L2.append(name_2[j])
        
# 2 : Remove the common characters betwins the tow names and get the count of the characters that are left :
    
    for a in L1[:]:            #[:] hendel with a coppy from the list
        if a in L2 :
            L1.remove(a)
            # print(L1)
            L2.remove(a)
            # print(L2)
            L3 = L1 + L2
            l = len(L3)
            
# 3 : Take FLAMES list :    

    F = ["Friends","Lovers","Admirers","Marriage","Enemies","Selebings"]       

# 4 : Removing list elment using the count we got (l):

    while (len(F) > 1):

        c = l%len(F)  # to find witch element we want to delet 

        s = c-1       # s : new index take the main index -1 (to start the count from 1)
        if s >= 0 :     
            left = F[:s]
            right = F[s+1:]
            F = right + left
                       
        else :
            F = F[:len(F)-1]   # if the len(F) == len(L3) (lengh of FLAMES list == number of characters that are left )
            
# 5 : The letter which lasts the process is the result :  

    print("You and your crash  : ",F[0])

Falme(name_1 , name_2)







#  f = [F , L , A , M , E , S]   
#  c = 0   1    2   3   4   5   
#  s = 1   2    3   4   5   6   s = c-1  
```
